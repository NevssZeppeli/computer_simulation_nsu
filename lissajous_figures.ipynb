```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = [1, 2, 60, 4]
B = [1, 6, 7, 8]
a = [2, np.sqrt(2), 30, 40]
b = [3, np.sqrt(np.pi), 150, 80]

fig, axs = plt.subplots(2, 2, figsize=(9, 8))
plt.subplots_adjust(wspace=0.6, hspace=0.2)

t = np.linspace(0, 100*np.pi, 1000)

fig.suptitle("Lissajous figures", fontsize=16)

axs[0,0].plot(A[0] * np.sin(a[0]*t + np.pi/2), B[0] * np.sin(b[0]*t), color="tab:green", lw=0.1)
axs[0,0].set_title(f"A={A[0]}, B={B[0]}, a={a[0]}, b={b[0]}")
axs[0,0].grid(True)

axs[0,1].plot(A[1] * np.sin(a[1]*t + np.pi/2), B[1] * np.sin(b[1]*t), color="tab:red", lw=0.2)
axs[0,1].set_title(f"A={A[1]}, B={B[1]}, a={round(a[1],4)}, b={round(b[1],4)}")
axs[0,1].grid(True)

axs[1,0].plot(A[2] * np.sin(a[2]*t + np.pi/2), B[2] * np.sin(b[2]*t), color="tab:orange", lw=0.2)
axs[1,0].set_title(f"A={A[2]}, B={B[2]}, a={a[2]}, b={b[2]}")
axs[1,0].grid(True)

axs[1,1].plot(A[3] * np.sin(a[3]*t + np.pi/2), B[3] * np.sin(b[3]*t), color="xkcd:eggplant", lw=0.4)
axs[1,1].set_title(f"A={A[3]}, B={B[3]}, a={a[3]}, b={b[3]}")
axs[1,1].grid(True)
```
